## 2. DataViz

Now, let's explore ways of understanding that data.

The first thing we should do is analyze the distribution of our variables. For each numeric variable, we can plot a simple `distribution plot`, a `boxplot`, a `stripplot` or a `violin plot`.

### 2.1 Distribution Plots

Let's start with the distribution plots:

```python
import seaborn as sns
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize = (15,6)) ##This line creates the ax we are going to use in order to plot our graphs
sns.distplot(data['area']) ##This creates a seaborn distribution plot that takes our variable as a single parameter
```

As we can see, the data must have some outliers, which makes it difficult to understand the distribution plot. One way to work around it is to use the `log` of the variable on the visualization:

```python
fig, ax = plt.subplots(figsize = (15,6)) ##This line creates the ax we are going to use in order to plot our graphs
sns.distplot(np.log(data['area'])) ## The np.log we used around our variable transforms it into its log version.
```

Now, let's analyze the same variable through other plots.

### 2.2 Boxplots

Boxplots allows us to understand the quantiles of our variable, and so, their distribution.

For the example below, we are using the log of the variable.

```python
# To make it easier to understand the code, I will create a column that is log of the 'area' variable, 
# so I don't have to keep typing the np.log on everyplot

data['area_log'] = np.log(data['area'])
```

The boxplot takes several arguments, such as values for the `x axis`, the `y axis`, and the `hue` argument, which helps us to see the same plot for different categories, as shown below:

```python
fig, ax = plt.subplots(figsize = (10,7)) ##Same old
sns.boxplot(x = data['area_log'], y = data['city'], hue = data['furniture'] ) 
```

### 2.3 Strip Plots

Strip plots is also a distribution representation, but it plots the data as a 1-dimension scatterplot. The parameters are very similar to the boxplot, but changing the `alpha` helps visualizing the distribution better.

```python
fig, ax = plt.subplots(figsize = (10,7)) ##Same old
sns.stripplot(x = data['area_log'], y = data['city'], hue = data['furniture'], alpha = 0.1 ) 
```

### 2.4 Violin Plot

The violin plot is somewhat the mix between the histogram and the box plot. It presents the information much like a boxplot, but it also shows how the distribution of the data behaves, not only its quantiles: 

```python
fig, ax = plt.subplots(figsize = (10,7)) ##Same old
sns.violinplot(x = data['area_log'], y = data['city'], hue = data['furniture'] ) 
```

### 2.5 QQ plot

The QQ Plot sorts your data into the theoretical quantiles, so you can see how it behaves from the minimum value to the maximum value.



```python
import statsmodels.api as sm
ax = sm.qqplot(data['area'])
```

As we can see, whe normal variable has some pretty big outliers. When we plot the log version of the variable, we have the following:

```python
ax = sm.qqplot(data['area_log'])
```

### 2.6 Multiple Plots 

Now, let's understand how we can plot several plots at the same time. This is really useful when we want to have a quick overview of our data.

```python
## First, we get all the numeric data we have:
from pandas.api.types import is_numeric_dtype

## This creates an empty list where we will put all of our numeric columns
numeric_cols = []

## Now, we scan our columns checking which one is numeric:
for col in data.columns:
    if is_numeric_dtype(data[col]):
        numeric_cols.append(col) ## If it is, we put it in that list
        
## Here we count how many plots we will have.
n_plots = len(numeric_cols)
        
## Here, we define the number of rows and columns of plots we want to have
n_cols = 3 
n_rows = int(round((n_plots)/n_cols+0.5,0)) ## We are defining the number of rows we want based on the number of cols.

fig = plt.subplots(figsize = (15,15)) # Same old, but no need for the 'ax'

for i, var in enumerate(numeric_cols):
    plt.subplot(n_rows,n_cols,i+1) # Here, defining the array of plots, with number of rows and columns, and the in what position our plot will be (i+1)
    plt.tight_layout(pad=2) ## Here we are defining the distance between the plots.
    sns.distplot(data[var])
    plt.title("Distribution of Variable: " + var)    
```

### 2.7 Categorical Variables 

Now, let's see some tools to better understand and visualize categorical variables. 

The count plot is a very good plot for that:

```python
fig, ax = plt.subplots(figsize = (10,7)) ##Same old
sns.countplot(data['city'])
```

Now, we do multiple plots for all the categorical variables:

```python
## First, we get all the categorical data we have:
from pandas.api.types import is_string_dtype

## This creates an empty list where we will put all of our numeric columns
string_cols = []

## Now, we scan our columns checking which one is numeric:
for col in data.columns:
    if is_string_dtype(data[col]):
        string_cols.append(col) ## If it is, we put it in that list
        
## Here we count how many plots we will have.
n_plots = len(string_cols)
        
## Here, we define the number of rows and columns of plots we want to have
n_cols = 2 
n_rows = int(round((n_plots)/n_cols+0.5,0)) ## We are defining the number of rows we want based on the number of cols.

fig = plt.subplots(figsize = (15,10)) # Same old, but no need for the 'ax'

for i, var in enumerate(string_cols):
    plt.subplot(n_rows,n_cols,i+1) # Here, defining the array of plots, with number of rows and columns, and the in what position our plot will be (i+1)
    plt.tight_layout(pad=2) ## Here we are defining the distance between the plots.
    sns.countplot(data[var])
    plt.title("Distribution of Variable: " + var)   
```
